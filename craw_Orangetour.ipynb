```python
# 建立dict用來轉換網站的欄位轉成自己的欄位
info_dict={'GrupCd': 'id',
           'GrupSnm':'travelName',
           'PortNm':'airPort',
           'LeavDt':'departureDate',
           'GrupLn':'days',
           'SaleAm':'price',
           'SaleYqt':'available',
           'EstmTotqt':'total',
           'ShareUrl':'detailUrl'}
```

```python
# 建立方法帶入網址就可取得該網站行程
def download_products(url,products=None):
    if not products:
        products=[]

    # 用迴圈跑需要的頁數
    for page in range(1, 5):
        print('url=',url,'pageAll=',page)
        # 抓存放行程資料的json檔
        f = session.post(url, data={'pageAll':page})
        f.encoding = 'utf-8'
        # 讀json檔, 行程內容都存放在All裡面, 把它們抓出來
        all_products = json.loads(f.text)['All']

        # 跑回圈取出每一個行程, 並把需要的欄位轉成我們自己的欄位取出來
        for all_product in all_products:
            product_dic={}
            for info in info_dict.keys():
                    product_dic[info_dict[info]] = all_product[info]
            
            products.append(product_dic)

        # 避免被網站伺服器偵測到
        time.sleep(5)
    
    return products
```

```python
url="http://www.orangetour.com.tw/EW/Services/SearchListData.asp"
products_orangetour=download_products(url)
```

```python
orangetour_ids = [product['id'] for product in products_orangetour]
print(len(set(orangetour_ids))==len(products_orangetour))
```
