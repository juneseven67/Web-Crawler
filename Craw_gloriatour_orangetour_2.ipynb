```python
import requests_html
import json, time, re
```

```python
session = requests_html.HTMLSession()
```

```python
# 建立dict用來轉換網站的欄位轉成自己的欄位
info_dict={'GrupCd': 'id',
           'GrupSnm':'name',
           'GrupLn':'days',
           'LeavDt':'date',
           'SaleAm':'price',
           'SaleYqt':'available',
           'EstmTotqt':'total'}
```

```python
#設置兩個網站的網址
urls=['https://www.gloriatour.com.tw/EW/Services/SearchListData.asp',
      'http://www.orangetour.com.tw/EW/Services/SearchListData.asp']
```

```python
# 建立方法帶入網址就可取得該網站行程
def download_products(url,products=None):
    if not products:
        products=[]

    # 用迴圈跑需要的頁數
    for page in range(1, 10):
        print('url=',url,'pageAll=',page)
        # 抓存放行程資料的json檔
        f = session.post(url, data={'pageALL':page})
        # 讀json檔, 行程內容都存放在All裡面, 把它們抓出來
        all_products = json.loads(f.text)['All']

        # 跑回圈取出每一個行程, 並把需要的欄位轉成我們自己的欄位取出來
        for all_product in all_products:
            product_dic={}
            for info in info_dict.keys():
                    product_dic[info_dict[info]] = all_product[info]
            
            products.append(product_dic)

        # 避免被網站伺服器偵測到
        time.sleep(5)
    
    return products
```

```python
products_gloriatour=download_products(urls[0])
products_orangetour=download_products(urls[1])
```

```python
# id是唯一且不重複的, 對它取set可以確保我們沒有抓到重複的資料
gloriatour_ids = [product['id'] for product in products_gloriatour]
print(len(set(gloriatour_ids))==len(products_gloriatour))
print('===================================')
orangetour_ids = [product['id'] for product in products_orangetour]
print(len(set(gloriatour_ids))==len(products_orangetour))
```

```python
for agency in [products_gloriatour,products_orangetour]:
    for product in agency:
        for info in product.keys():
             print(str(info), str(product[info]), sep=' : ')
        print('==========================')
    print('products_gloriatour end,products_orangetour start!')
```

```python
def examination(url):
# 抓下網頁第一頁的資料
    f = session.get(url)
    # 抓網頁上的行程
    thumbnails = f.html.find('.thumbnail', first=False)
    # 驗證資料是否與網頁相同,印出代表正確
    for product in products_gloriatour:
        print(product['id'], product['name'])

        for thumbnail in thumbnails:
            if not re.search(product['id'], thumbnail.find('.product_num', first=True).text):
                continue

            if not re.search(product['date'], thumbnail.find('.product_date', first=True).text):
                raise Exception(product['date']+thumbnail.find('.product_date', first=True).text)

            if not re.search(str(product['days']), thumbnail.find('.product_days', first=True).text):
                raise Exception(str(product['days'])+thumbnail.find('.product_days', first=True).text)

            if not re.search(format(product['price'],','), thumbnail.find('.product_price', first=True).text):
                raise Exception(format(product['price'],',')+thumbnail.find('.product_price', first=True).text)

            if not re.search(str(product['available']), thumbnail.find('.product_available', first=True).text):
                raise Exception(str(product['available'])+thumbnail.find('.product_available', first=True).text)

            if not re.search(str(product['total']), thumbnail.find('.product_total', first=True).text):
                raise Exception(str(product['total'])+thumbnail.find('.product_total', first=True).text)
```

```python
examination(urls_2[0])
print('============products_gloriatour check ok!==============')
examination(urls_2[1])
print('============products_orangetour check ok!==============')
```

#### 兩個網站的處理

```python
import pandas as pd
```

```python
#建立兩個個字dataframe,加入旅行社名稱欄位
gloriatour_df=pd.DataFrame(data=products_gloriatour,columns=['id','date','name','days','price','total','available'])
gloriatour_df['travel_agency']='gloriatour'

orangetour_df=pd.DataFrame(data=products_orangetour,columns=['id','date','name','days','price','total','available'])
orangetour_df['travel_agency']='orangetour'
```

```python
total_df=pd.concat([gloriatour_df,orangetour_df],axis=0)
```

```python
#做欄位順序
col=['id','date','travel_agency']
col2=list(set(gloriatour_df.columns)-set(col))
col2.sort(key=list(gloriatour_df.columns).index)
total_df_new=pd.concat([total_df.loc[:,col],total_df.loc[:,col2]],axis=1)
#照date,travel_agency,name排序
total_df_new.sort_values(by=['date','travel_agency','name'],inplace=True)
#索引重排
total_df_new=total_df_new.reset_index().drop("index",axis=1)
```

```python
#輸出csv檔
total_df_new.to_csv('travel_products.csv',index=False,encoding='UTF-8')
```
